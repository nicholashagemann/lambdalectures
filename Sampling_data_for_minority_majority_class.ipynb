```python
import pandas as pd
```

```python
# Sample data
X = pd.DataFrame({'num_legs': [2, 4, 8, 0,3,5,8,9,2],
                   'num_wings': [2, 0, 0, 0,1,4,2,5,6],
                   'num_specimen_seen': [10, 2, 1, 8,28,22,1,3,5],
                   'label':[1,1,0,0,1,1,0,0,1]},
                  index=['falcon', 'dog', 'spider', 'fish','cat', 'mouse',' rhino','pig', 'cow'])
```

```python
X
```

```python
# Splitting data based on labels
X_true = X[X['label'] == 1]
X_false = X[X['label'] == 0]
```

```python
X_true
```

```python
X_false
```

```python
# When you over sample, you always have to sample with replacement
X_true_samp = X.sample(20,replace = True)
```

```python
X_true_samp
```

```python
# Under sampling doesnt need with replacement
X_false_samp = X_false.sample(3,replace=False)
```

```python
X_false_samp
```

```python
X_final = X_true_samp.append(X_false_samp, ignore_index=True)
```
